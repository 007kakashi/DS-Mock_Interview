## R Decision Making

Decision Making statements help to decide whether to execute a block of code or not, based on a condition. Decision making is an important aspect in any programming language. Decision Making statements are also called Conditional statement.

### R If Statement

R If statement executes a set of statements based on a given condition. If the condition in If-statement evaluates to TRUE, then the statements in the If block execute, else not.



#### Syntax

```r
a <- 6
 
if (a == 6) {
  print ("Value in a is 6.")
}
```

```r
# Check if Number is Even using If
n <- 10
 
if (n %% 2 == 0) {
  print ("Number is even.")
}
```

### R If Else

#### Syntax 

```r
# for TRUE condition
a = 6
 
if(a==6){
    print ("Condition a==6 is TRUE")
    print ("This is second statement in if block")
} else{
    print ("Condition a==6 is FALSE")
    print ("This is second statement in else block")
} 
```

```r
# for FALSE condition
b = 7
 
if(b==6){
    print ("Condition b==6 is TRUE")
    print ("This is second statement in if block")
} else{
    print ("Condition b==6 is FALSE")
    print ("This is second statement in else block")
}
```

### Nested If Else Statement

```r
x <- 5
if (x == 0) {
  print("Zero")
} else if (x > 0) {
  print("Positive number")
} else print("Negative Number")
```

```r
if (x>=0){
  if (x==0){
    print("Number is Zero")
  }else print("Number is Positive")
}else print("Number is negative")
```

```r
# R if...else if...else statement Example
 
b = 7
 
if(b==6){
    print ("Condition b==6 is TRUE")
} else if(b==7){
    print ("Condition b==7 is TRUE")
} else if(b==8){
    print ("Condition b==8 is TRUE")
} else{
    print ("No if condition is TRUE for b")
}
```

## R Switch

R switch statement selects one of the cases, based on the value of an expression.

#### Switch based on Index

```r
y <- 3
 
x <- switch(
    y,
    "Good Morning",
    "Good Afternoon",
    "Good Evening",
    "Good Night"
)
 
print(x)
```

#### Switch based on Matching Value

```r
y <- 3
 
x <- switch(
    y,
    "Good Morning",
    "Good Afternoon",
    "Good Evening",
    "Good Night"
)
 
print(x)
```

```r
y <- "12"
 
x <- switch(
    y,
    "9"="Good Morning",
    "12"="Good Afternoon",
    "18"="Good Evening",
    "21"="Good Night"
)
 
print(x)
```

# R Loop Statements

Repeating execution of a block of statements in a controlled way is an important aspect in any functional programming language. It helps to keep the code concise and clean.



R programming language provides three looping statements. They are :
- **R repeat loop**  A block of statements are repeated forever. A breaking condition has to be provided inside the repeat block to come out of the loop.

- **R while loop**  A block of statements are executed repeatedly in a loop till the condition provided to while statement returns TRUE.

- **R for loop**  A block of statements are executed for each of the items in the list provided to the for loop.


## Repeat Loop in R

R Repeat Loop does the execution of a set of statements in a loop until a break condition is met. In this R Tutorial, we shall learn Repeat Loop syntax and execution flow

```r
# R Repeat Loop statement Example
 
a = 1
 
repeat {
    # starting of repeat statements block
    print(a)
    a = a+1
    # ending of repeat statements block
    if(a>6){ # breaking condition
        break
    }
}
```

## While Loop

R While loop has a condition and a body. While Loop executes a set of statements in the body repeatedly in a loop as long as the condition is satisfied.

#### A Simple While Loop to Print Numbers from 1 to N

```r
a <- 1
 
while(a<5) {
  print(a)
  a = a+1
}
```

```r
a <- 1
b <- 3
 
while(a<5) {
  if (a == b) {
    break
  }
  print(a)
  a = a+1
}
```

## R For Loop 
R For Loop executes a set of statements for each of the elements in a vector provided to it.

```r
# R for loop Example
 
a = c(2, 45, 9, 12)
 
for(i in a) {
    print(i*2)
}
```

```r
# R for loop example with break statement
 
a = c(12, 45, 9, 12)
 
for(i in a) {
    if(i==9){
        break
    }
    print(i)
}
```

# Functions in R

A function is a set of instructions to perform a specific task. In a programming language, a function transforms the arguments provided into a desired form.

For example, consider a function that does addition. It calculates the sum of provided arguments. The calculation part is the transformation and final sum is the desired form of arguments.

#### Syntax

```r
# R function with no arguments
sayHello = function(){
    print("Hello !")
}
 
sayHello()
```

```r
# R function with arguments
addition = function(a,b,c){
    print(a+b+c)
}
 
addition(4,15,6)
```

```r
# we will write a function that returns a value. We will store this returned value in a variable, say d.
# R function with arguments and return value
addition = function(a,b,c){
    return (a+b+c)
}
 
d = addition(4,15,6)
 
print(d)
```
